## Standardization
 We try to bring all the variables or features to a simillar scale. Standardisation means centering variables at zeros. z = (x - x_mean)/std

```python
import pandas as pd
df = pd.read_csv('titanic_train.csv', usecols=['Pclass','Age','Fare','Survived'])
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df.Age.median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.33, 
                                                    random_state = 42)
```

```python
scaler = StandardScaler()
##fit vs fit_transform
df_scaled = scaler.fit_transform(df)
df_scaled
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled
```

```python
scaler.transform(X_test)
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
##Model Building
##fit() for training data and predict() for test data
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression()
```

```python
classification.fit(X_train_scaled, y_train)
```

```python
classification.predict(X_test_scaled)
```
